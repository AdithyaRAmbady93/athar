# Handling Outliers

```python
# Visualize outliers using box plots
plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.boxplot(x='LoanAmount', data=train, palette='Set1')
plt.title('LoanAmount Boxplot')

plt.subplot(122)
sns.boxplot(x='ApplicantIncome', data=train, palette='Set1')
plt.title('ApplicantIncome Boxplot')

plt.show()
```

```python
# Define a function to treat outliers using a capping method
def treat_outliers(data, column, multiplier=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Treat outliers for 'LoanAmount' and 'ApplicantIncome'
treat_outliers(train, 'LoanAmount')
treat_outliers(train, 'ApplicantIncome')

# Check if outliers are treated
plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.boxplot(x='LoanAmount', data=train, palette='Set1')
plt.title('LoanAmount Boxplot after Outlier Treatment')

plt.subplot(122)
sns.boxplot(x='ApplicantIncome', data=train, palette='Set1')
plt.title('ApplicantIncome Boxplot after Outlier Treatment')

plt.show()
```

# Model Building

```python
train=train.drop('Loan_ID',axis=1)
test=test.drop('Loan_ID',axis=1)
```

```python
#Separate Target Variable
X = train.drop('Loan_Status',1)
y = train.Loan_Status
```

```python
#Convert Categorical Variables to Dummy Variables:
X=pd.get_dummies(X)
train=pd.get_dummies(train)
test=pd.get_dummies(test)
```

```python
#Split Data into Training and Validation Sets
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(X,y, test_size =0.3,random_state=42)
```

```python
#Train Logistic Regression Model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Loan_Status for the validation set and calculate its accuracy.

pred_cv = model.predict(x_cv)

# calculating the accuracy.
accuracy_score(y_cv,pred_cv)
```

```python
#predictions for the test dataset
pred_test = model.predict(test)
```
